```python
import pandas as pd

df = pd.read_csv('Inventory Management E-Grocery - InventoryData.csv')


# Received_Date, Last_Purchase_Date, Expiry_Date, Audit_Date should be datetime types
# Avg_Daily_Sales, Days_of_Inventory, SKU_Churn_Rate, Order_Frequency_per_month should be float (change , to .)
# Unit_Cost_USD, Last_Purchase_Price_USD, Total_Inventory_Value_USD should be float (delete $ and change , to .)
# Supplier_OnTime_Pct, Audit_Variance_Pct, Demand_Forecast_Accuracy_Pct should be float (delete %)

df['Received_Date'] = pd.to_datetime(df['Received_Date'])
df['Last_Purchase_Date'] = pd.to_datetime(df['Last_Purchase_Date'])
df['Expiry_Date'] = pd.to_datetime(df['Expiry_Date'])
df['Audit_Date'] = pd.to_datetime(df['Audit_Date'])

df['Avg_Daily_Sales'] = df['Avg_Daily_Sales'].str.replace('.', '', regex=False).str.replace(',', '.').astype(float)
df['Days_of_Inventory'] = df['Days_of_Inventory'].str.replace('.', '', regex=False).str.replace(',', '.').astype(float)
df['SKU_Churn_Rate'] = df['SKU_Churn_Rate'].str.replace('.', '', regex=False).str.replace(',', '.').astype(float)
df['Order_Frequency_per_month'] = df['Order_Frequency_per_month'].str.replace('.', '', regex=False).str.replace(',', '.').astype(float)

df['Unit_Cost_USD'] = df['Unit_Cost_USD'].str.replace('$', '').str.replace('.', '', regex=False).str.replace(',', '.').astype(float)
df['Last_Purchase_Price_USD'] = df['Last_Purchase_Price_USD'].str.replace('$', '').str.replace('.', '', regex=False).str.replace(',', '.').astype(float)
df['Total_Inventory_Value_USD'] = df['Total_Inventory_Value_USD'].str.replace('$', '').str.replace('.', '', regex=False).str.replace(',', '.').astype(float)
df['Supplier_OnTime_Pct'] = df['Supplier_OnTime_Pct'].str.replace('%', '').str.replace(',', '.').astype(float)
df['Audit_Variance_Pct'] = df['Audit_Variance_Pct'].str.replace('%', '').str.replace(',', '.').astype(float)
df['Demand_Forecast_Accuracy_Pct'] = df['Demand_Forecast_Accuracy_Pct'].str.replace('%', '').str.replace(',', '.').astype(float)
df
```

```python
#1) Сгруппируйте по категориям (Category) и найдите 3 категории с наибольшей суммой Total_Inventory_Value_USD.

df.groupby("Category")["Total_Inventory_Value_USD"].sum().nlargest(3)

#Вариант 2
#Average_age = df_1.groupby('Category')['Stock_Age_Days'].mean().round(2)
#Average_age.sort_values(ascending= False)[:3]
```

```python
#2) Вычислите средний возраст запасов для каждой категории. (группировка)

df.groupby("Category")["Stock_Age_Days"].mean()
```

```python
#3) Добавьте новую колонку Stock_Turnover_Rate (формула Avg_Daily_Sales * 365 / Quantity_On_Hand) (чтобы понять обороты товаров)

df["Stock_Turnover_Rate"] = round((df["Avg_Daily_Sales"] * 365) / (df["Quantity_On_Hand"]),2)
df
```
