```python
#1.
#Создайте три переменные: name, age, city. 

#Выведите фразу:
#Меня зовут Ali, мне 20 лет, я живу в Baku.



name = "Ali"
age = 20
city = "Baku"
print(f"Меня зовут {name}, мне {age}, я живу {city}")
```

```python
#2.

#Пользователь вводит число.

#Если оно больше нуля → вывести "Положительное".
#Если меньше нуля → "Отрицательное".
#Если равно нулю → "Ноль".

num = float(input("Введите число: "))

if num > 0:
    print("Положительное")
elif num == 0:
    print("Ноль")
elif num < 0:
    print("Отрицательное")
```

```python
#3.
#Дан список чисел:
#nums = [2, 5, 7, 10]
#3.1 Найдите сумму чисел с помощью цикла.
#3.2 Найдите сумму чисел с помощью специальной функции.

nums = [2, 5, 7, 10]
sum_n = 0
for num in nums:
    sum_n += num
print(f"Сумма через цикл: {sum_n}")

function_sum = sum(nums)
print(f"Сумма через функцию: {function_sum}")
```
